```python
import pandas as pd
import numpy as np
path = "/content/netflix_titles.csv"
df = pd.read_csv(path)

print("Shape:", df.shape)
df.head()
```

```python
df.info()
df.describe(include="all")

```

```python
print("Before:", df.shape)
df = df.drop_duplicates()
print("After:", df.shape)
```

```python
print("Missing values before:\n", df.isnull().sum())

# Fill categorical missing values with "Unknown"
for col in ["director", "cast", "country"]:
    if col in df.columns:
        df[col] = df[col].fillna("Unknown")

# Drop rows missing date_added or rating (small number of rows)
df = df.dropna(subset=["date_added", "rating"])

print("\nMissing values after:\n", df.isnull().sum())
```

```python
df.columns = (
    df.columns.str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

df = df.rename(columns={
    "show_id": "id",
    "listed_in": "genre"
})

print(df.columns.tolist())
```

```python

cols_to_drop = ["description"]
df = df.drop(columns=[c for c in cols_to_drop if c in df.columns])

print(df.columns.tolist())
```

```python
print("Final shape:", df.shape)
df.info()
df.describe(include="all")
```

```python
output_path = r"C:\Users\PC\OneDrive\Documents\netflix_titles_cleaned.csv"
df.to_csv(output_path, index=False)
print("Saved cleaned file to:", output_path)
```
