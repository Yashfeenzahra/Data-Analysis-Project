```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/netflix_titles.csv")
df["date_added"] = pd.to_datetime(df["date_added"], errors="coerce")

print("Shape:", df.shape)
df.head()
```

Total Records


```python
total_records = df.shape[0]
unique_titles = df["title"].nunique()

print(f"Total records: {total_records}")
print(f"Unique titles: {unique_titles}")

if total_records != unique_titles:
    print("Note: some titles repeat in the dataset.")
else:
    print("All titles are unique.")
```

Highest and Lowest Values

```python

df["duration_value"] = df["duration"].str.extract(r'(\d+)').astype(float)

movies = df[df["type"] == "Movie"]
tv = df[df["type"] == "TV Show"]

longest_movie = movies.loc[movies["duration_value"].idxmax()]
shortest_movie = movies.loc[movies["duration_value"].idxmin()]
most_seasons = tv.loc[tv["duration_value"].idxmax()]

print(f"Longest movie: {longest_movie['title']} ({longest_movie['duration_value']:.0f} min)")
print(f"Shortest movie: {shortest_movie['title']} ({shortest_movie['duration_value']:.0f} min)")
print(f"TV show with most seasons: {most_seasons['title']} ({most_seasons['duration_value']:.0f} seasons)")
```

Avg Calc

```python
avg_movie_duration = movies["duration_value"].mean()
avg_tv_seasons = tv["duration_value"].mean()

print(f"Average movie duration: {avg_movie_duration:.1f} minutes")
print(f"Average number of seasons per TV show: {avg_tv_seasons:.2f}")

```

Category Wise Analysis

```python
type_counts = df["type"].value_counts()
print("Titles by type:\n", type_counts)
```

```python
rating_counts = df["rating"].value_counts()
print("Titles by rating:\n", rating_counts)
```

```python
country_series = df["country"].str.split(", ").explode()
top_countries = country_series.value_counts().head(10)
print("Top 10 producing countries:\n", top_countries)
```

Trend Analysis


```python
titles_per_year = df["date_added"].dt.year.value_counts().sort_index()
print("Titles added to Netflix, per year:\n", titles_per_year)
```

```python
recent_year = int(df["date_added"].dt.year.max())
monthly = df[df["date_added"].dt.year == recent_year]["date_added"].dt.month.value_counts().sort_index()
print(f"Titles added by month in {recent_year}:\n", monthly)
```

Correlation ANalysis


```python
movie_corr = movies[["release_year", "duration_value"]].corr().iloc[0, 1]
tv_corr = tv[["release_year", "duration_value"]].corr().iloc[0, 1]

print(f"Correlation (release_year vs movie duration): {movie_corr:.3f}")
print(f"Correlation (release_year vs TV seasons): {tv_corr:.3f}")
print("\nA value near 0 means little to no relationship; closer to 1 or -1 means a stronger trend.")
```

Most Freq Values


```python
genre_series = df["listed_in"].str.split(", ").explode()
top_genres = genre_series.value_counts().head(10)
print("Top 10 most frequent genres:\n", top_genres)
```

```python
director_counts = df[df["director"] != "Unknown"]["director"].value_counts().head(10)
print("Top 10 most frequent directors (excluding unknown):\n", director_counts)
```
